#### `Probability Ratio Encoding`

1. Probability of Survived based on Cabin -- Categorical Features
2. Probability of Not Survived -- 1 - pr(Survived)
3. pr(Survived)/pr(Not Survived)
4. Dictionary to map Cabin with probability
5. Replace / map with the categorical feature

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/krishnaik06/Feature-Engineering-Live-sessions/master/titanic.csv'
```

```python
df = pd.read_csv(url, usecols=['Cabin', 'Survived'])
df.head()
```

```python
### Replacing NAN values with `Missing`

df['Cabin'].fillna('Missing', inplace=True)
df.head()
```

```python
df.head(20)
```

```python
df['Cabin'].unique()
```

```python
df['Cabin'] = df['Cabin'].astype(str).str[0]
```

```python
df.head()
```

```python
df.Cabin.unique()
```

```python
### Check the missing value %
df.groupby(['Cabin'])['Survived'].mean()*100
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
prob_df = df.groupby(['Cabin'])['Survived'].mean()
```

```python
prob_df = pd.DataFrame(prob_df)
prob_df
```

```python
prob_df['Died'] = 1 - prob_df['Survived']
```

```python
prob_df.head()
```

```python
prob_df['Probability_ratio'] = prob_df['Survived']/prob_df['Died']
```

```python
prob_df.head()
```

```python
prob_df['Probability_ratio'].to_dict()
```

```python
probability_encoded = prob_df['Probability_ratio'].to_dict()
```

```python
df['Cabin_encoded']=df['Cabin'].map(probability_encoded)
```
