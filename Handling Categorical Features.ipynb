### `One Hot Encoding with many categories in a feature`

```python
mercedes_data = "https://raw.githubusercontent.com/krishnaik06/Feature-Engineering-Live-sessions/master/mercedes.csv"
```

```python
df = pd.read_csv(mercedes_data)
```

```python
df.head()
```

```python
df = pd.read_csv(mercedes_data, usecols=['X0','X1','X2','X3','X4','X5','X6'])
```

```python
df.head()
```

```python
df['X1'].value_counts()
```

```python
df['X0'].unique()
```

```python
for i in df.columns:
  print(len(df[i].unique()))
```

```python
#### Take the top 10 most frequent categories

df.X1.value_counts().sort_values(ascending=False).head(10)
```

```python
lst_10 = df.X1.value_counts().sort_values(ascending=False).head(10).index
lst_10=list(lst_10)
```

```python
lst_10
```

```python
import numpy as np
for categories in lst_10:
  df[categories] = np.where(df['X1'] == categories,1,0)
```

```python
df[lst_10]
```

```python
lst_10.append('X1')
```
